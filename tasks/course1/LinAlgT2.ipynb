```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
%matplotlib inline  

def f(x):
    a = np.sin(x/5.0) * np.exp(x/10.0) 
    a += 5.0 * np.exp(-x / 2.0)
    return a

def fa(x, x1):
    res = 0
    n,_ = x1.shape
    for i in xrange(n):
        res += x1[i, 0] * (x ** i)
    return res

t1 = np.arange(1, 15.1, 0.1)
print f(1.0)
plt.plot(t1, f(t1))
plt.show()

A1 = np.array([
    [1, 1],
    [1, 15]
])

b1 = np.array([[f(1)], [f(15)]])
x1 = solve(A1, b1)
print '1st'
plt.plot(t1, f(t1))
plt.plot(t1, fa(t1, x1))
plt.show()
A2 = np.array([
    [1, 1, 1],
    [1, 8, 64],
    [1, 15, 225]
])
b2 = np.array([
    [f(1.0)],
    [f(8.0)],
    [f(15.0)]
])
x2 = solve(A2, b2)
print '2nd'
plt.plot(t1, f(t1))
plt.plot(t1, fa(t1, x2))
plt.show()

A3 = np.array([
    [1, 1, 1, 1],
    [1, 4, 16, 64],
    [1, 10, 100, 1000],
    [1, 15, 225, 15 ** 3]
])
b3 = np.array([
    [f(1)],
    [f(4)],
    [f(10)],
    [f(15)]
])
x3 = solve(A3, b3)
print "3rd"

plt.plot(t1, f(t1))
plt.plot(t1, fa(t1, x3))
plt.show()


with open('submission-2.txt', 'w') as f:
    f.write(' '.join(str(x) for x in x3[:, 0].tolist()))

%cat submission-2.txt
```
